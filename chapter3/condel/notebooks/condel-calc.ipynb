The HumVar datasets use Swissprot IDs for the proteins but the FannsDB use Ensembl protein IDs so I will need these maps to do the queries in the database. The tool *ensembl-get-ann* makes a query to the Ensembl biomart system for the Human genes dataset with two attributes (passed as parameters) for the chromosomes 1-22,X,Y and remove duplicated rows and rows with blank cells. This is the code of the bash script:

```python
%%file ensembl-get-ann
#!/bin/bash

if [ $# -lt 4 ]; then
	echo "Usage: ensembl-get-ann <host> <file> <attr-src> <attr-dst>"
	exit -1
fi

host=$1
file=$2
attr_src=$3
attr_dst=$4

echo
echo "From $host: $attr_src * $attr_dst --> $file ..."
echo

date "+# %D %T" >$file
echo "## host=$host" >>$file
echo "## attr_src=$attr_src" >>$file
echo "## attr_dst=$attr_dst" >>$file

(curl -o - --data-urlencode query@-  http://$host/biomart/martservice |
 gawk '$1 != "" && $2 != ""' |
 grep -vE "^LRG.*" |
 sort -u >>$file) <<EOF
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE Query>
<Query  virtualSchemaName = "default" formatter = "TSV" header = "0" uniqueRows = "1" count = "" datasetConfigVersion = "0.6" >	
	<Dataset name = "hsapiens_gene_ensembl" interface = "default" >
		<Attribute name = "$attr_src" />
		<Attribute name = "$attr_dst" />
	</Dataset>
</Query>
EOF
```

## Fetch scores

The following function fetches the scores from the database from a protein position, takes care of the mapping between Swissprot IDs and Ensembl IDs and the longest transcript.

```python
%%px --local
def fetch_scores(ds_path, out_path, predictors, trs_len, prot_len, swp_map, db_conf):
    
    find_fields = {"s." + pred : 1 for pred in predictors}
    find_fields.update({"g.t" : 1})
    
    with MongoClient(db_conf["url"]) as conn:
        db = conn[db_conf["db"]]
        with open(ds_path) as df, open(out_path, "w") as of:
            of.write("\t".join(["PROTEIN", "AA_POS", "AA_REF", "AA_ALT"] + predictors) + "\n")
            num_lines = unmapped = no_scores = 0
            for num_lines, line in enumerate(df, start=1):
                fields = line.rstrip("\n").split("\t")
                swpid, apos, aref, aalt = fields[0:4]
                if swpid not in swp_map:
                    unmapped += 1
                    continue

                # we are only interested in the scores for the longest transcript/protein
                ensp = reduce(
                    lambda p1,p2: p1 if prot_len[p1] >= prot_len[p2] else p2, swp_map[swpid])

                # Get the scores for the longest transcript/protein
                scores = { name : (0, None) for name in predictors }
                for d in db.scores.find({"p.n" : ensp, "p.p" : int(apos), "p.r" : aref, "p.a" : aalt}, find_fields):

                    trs_ln = trs_len.get(d["g"]["t"], 0)
                    for pred, score in d["s"].items():
                        prev_len, prev_score = scores[pred]
                        if trs_ln >= prev_len:
                            scores[pred] = (trs_ln, score)

                if sum([1 if s is not None else 0 for l, s in scores.values()]) == 0:
                    no_scores += 1
                    continue
                
                sb = [ensp, apos, aref, aalt] + [scores[pred][1] for pred in predictors]
                of.write("\t".join([str(v) if v is not None else "" for v in sb]) + "\n")

    return num_lines, unmapped, no_scores
```

## Functions needed to calculate Condel scores

### Single SNV calculation

This function calculates condel for a single SNV:

```python
%%px --local
def condel_calc_snv(snv, transforms, metrics, predictors):
    import numpy as np
    
    scores = snv["s"]
    condel = wsum = 0
    for pred in predictors:
        if pred not in scores:
            continue

        score = scores[pred]
        if pred in transforms:
            score = transforms[pred](score)
            
        pm = metrics[pred]
        rmin, rmax = pm["rmin"], pm["rmax"]
        cutoff = pm["best_perf_cutoff"]["MCC"]
        pbins = np.array(pm["pos_bins"])[:-1]
        nbins = np.array(pm["neg_bins"])[:-1]
        ncd, pcd = pm["neg_cumdist"], pm["pos_cumdist"]
        
        if score < cutoff:
            idx = np.argmin(np.abs(nbins - score))
            w = 1 - ncd[idx]
        else:
            idx = np.argmin(np.abs(pbins - score))
            w = pcd[idx]
        
        norm_score = (score - rmin) / (rmax - rmin)
        condel += w * norm_score
        wsum += w
    
    if wsum == 0:
        return None
    
    return condel / wsum
```

### Calculate Condel scores only for the performance datasets

```python
%%px --local
def condel_calc_ds(ds_path, out_path, predictors, trs_len, prot_len, swp_map, db_conf):
    import json
    from pymongo import MongoClient
    
    find_fields = {"s." + pred : 1 for pred in predictors}
    find_fields.update({"g" : 1, "p" : 1})
    
    with open(METRICS_PATH) as f:
        metrics = json.load(f)["metrics"]
    
    with MongoClient(db_conf["url"]) as conn:
        db = conn[db_conf["db"]]
        with open(ds_path) as df, open(out_path, "w") as of:
            of.write("\t".join(["PROTEIN", "AA_POS", "AA_REF", "AA_ALT"] + predictors + ["CONDEL"]) + "\n")
            num_lines = unmapped = no_scores = 0
            for num_lines, line in enumerate(df, start=1):
                fields = line.rstrip("\n").split("\t")
                swpid, apos, aref, aalt = fields[0:4]
                if swpid not in swp_map:
                    unmapped += 1
                    continue

                # we are only interested in the scores for the longest transcript/protein
                ensp = reduce(
                    lambda p1,p2: p1 if prot_len[p1] >= prot_len[p2] else p2, swp_map[swpid])

                # Get the scores for the longest transcript/protein
                last_len = 0
                for d in db.scores.find({"p.n" : ensp, "p.p" : int(apos), "p.r" : aref, "p.a" : aalt}, find_fields):

                    trs_ln = trs_len.get(d["g"]["t"], 0)
                    if trs_ln <= last_len:
                        continue
                    
                    last_len = trs_ln
                    
                    condel = condel_calc_snv(d, TRANSFORMS, metrics, CONDEL_PREDICTORS)
                    
                    if condel is None:
                        no_scores += 1
                        continue
                    
                    s = d["s"]
                    del d["s"]
                    s["CONDEL"] = condel
                    db.condel.update(d, {"$set" : {"s" : s}}, upsert=True)
                    
                    p = d["p"]
                    pred_scores = [str(s[pred]) if pred in s else "" for pred in predictors] + [str(condel)]
                    of.write("\t".join([p["n"], str(p["p"]), p["r"], p["a"]] + pred_scores) + "\n")

    return num_lines, unmapped, no_scores

def condel_calc_all_ds(prefix):
    from datetime import datetime as dt
    
    print("{} Calculating CONDEL for {} ...".format(dt.now(), prefix))
    
    num_lines, unmapped, no_scores = condel_calc_ds(
        os.path.join(WORK_PATH, prefix),
        os.path.join(WORK_PATH, "{}.condel".format(prefix)),
        PREDICTORS, trs_len, prot_len, swp_map, FANNSDB)
    
    print("{} num_lines={}, unmapped={}, no_scores={} ...".format(dt.now(), num_lines, unmapped, no_scores))
```

### Plot the ROC curve

```python
from sklearn.metrics import matthews_corrcoef, roc_curve, auc

class CondelModel(object):
    '''This is an object oriented implementation of condel.metrics.perf_metrics
    more convenient for the cross-validation'''
    
    def __init__(self, X, y):
        self.X = X
        self.y = y
        
        self.metrics = {}
        self.cutoff = np.nan
        
        self.__calc_model()
        self.__calc_cutoff()
    
    def __calc_metrics(self, X, y):
        d = pd.DataFrame({"X" : X, "y" : y})
        d = d[np.isfinite(X)]
        if len(d) == 0:
            return {}
        
        d_pred = d["X"].values
        y_true = d["y"].values
        rmin = d_pred.min()
        rmax = d_pred.max()
        
        dw = 1 / 64.0
        cutoffs = np.arange(rmin, rmax + dw, dw)
        mcc = np.empty(len(cutoffs))
        for i, cutoff in enumerate(cutoffs):
            y_pred = np.array(d_pred >= cutoff, dtype=np.int8)
            mcc[i] = matthews_corrcoef(y_true, y_pred)

        cutoff = cutoffs[np.argmax(mcc)]

        y_pred = np.array(d_pred >= cutoff, dtype=np.int8)
        fpr, tpr, thresholds = roc_curve(y_true, d_pred)
        roc = pd.DataFrame({"FPR" : fpr, "TPR" : tpr, "thresholds" : thresholds})
        auc_value = auc(fpr, tpr)

        n = d[d.y == 0]["X"].values
        p = d[d.y == 1]["X"].values

        nhist, nbins = np.histogram(n, bins=cutoffs)
        ncumhist = np.cumsum(nhist)
        ncummax = float(ncumhist[-1])
        ncumdist = ncumhist / ncummax
        ncutoffidx = np.argmin(np.abs(nbins[:-1] - cutoff))

        phist, pbins = np.histogram(p, bins=cutoffs)
        pmax = float(phist.max())
        pcumhist = np.cumsum(phist)
        pcummax = float(pcumhist[-1])
        pcumdist = pcumhist / pcummax
        pcutoffidx = np.argmin(np.abs(pbins[:-1] - cutoff))

        return dict(
            rmin=rmin, rmax=rmax,
            cutoffs=cutoffs, mcc=mcc, cutoff=cutoff,
            roc=roc, auc=auc_value,            
            neg_hist=nhist,
            neg_bins=nbins,
            neg_cumhist=ncumhist,
            neg_cumdist=ncumdist,
            neg_cutoff_index=ncutoffidx,
            pos_hist=phist,
            pos_bins=pbins,
            pos_cumhist=pcumhist,
            pos_cumdist=pcumdist,
            pos_cutoff_index=pcutoffidx
        )
        
    def __calc_model(self):
        for pi, pred in enumerate(self.X):
            self.metrics[pred] = self.__calc_metrics(self.X[pred], self.y)
    
    def __predict_case(self, scores, predictors):
        condel = wsum = 0
        for pred in predictors:
            score = scores[pred]
            if not np.isfinite(score):
                continue

            pm = self.metrics[pred]
            rmin, rmax = pm["rmin"], pm["rmax"]
            cutoff = pm["cutoff"]
            pbins = pm["pos_bins"][:-1]
            nbins = pm["neg_bins"][:-1]
            ncd = pm["neg_cumdist"]
            pcd = pm["pos_cumdist"]

            if score < cutoff:
                idx = np.argmin(np.abs(nbins - score))
                w = 1 - ncd[idx]
            else:
                idx = np.argmin(np.abs(pbins - score))
                w = pcd[idx]

            norm_score = (score - rmin) / (rmax - rmin)
            condel += w * norm_score
            wsum += w

        if wsum == 0:
            return np.nan

        return condel / wsum
    
    def __calc_cutoff(self):
        X = self.predict_score(self.X)
        m = self.__calc_metrics(X, self.y)
        self.metrics["CONDEL"] = m
        self.cutoff = m["cutoff"]
        
    def predict_score(self, x):
        return x.apply(partial(self.__predict_case, predictors=x.columns), axis=1)
    
    def predict(self, x):
        y = x.apply(partial(self.__predict_case, predictors=x.columns), axis=1)
        return pd.DataFrame({"y" : y, "del" : y > self.cutoff})
```

### Per protein calculation

This function calculates condel for all the SNVs of a given protein:

```python
%%px --local
def condel_calc_protein(protein, metrics_path, predictors, db_conf):
    import json
    from pymongo import MongoClient

    task = Task("condel-calc", protein, db_conf)
    
    task.debug("MongoDB configuration: {}".format(db_conf))
    
    task.info("Loading metrics ...")
    
    with open(metrics_path) as f:
        metrics = json.load(f)["metrics"]

    task.info("Calculating ...")
    
    with MongoClient(db_conf["url"]) as conn:
        db = conn[db_conf["db"]]
        num_docs = 0
        updates = 0
        for d in db.scores.find({"p.n" : protein}, {"s.{}".format(p) : 1 for p in predictors}):
            num_docs += 1
            condel = condel_calc_snv(d, TRANSFORMS, metrics, predictors)
            if condel is not None:
                updates += 1
                db.scores.update({"_id" : d["_id"]}, {"$set" : {"s.CONDEL" : condel}})
            if num_docs % 1000 == 0:
                task.progress(num_docs, "num_docs={}, updated_docs={}".format(num_docs, updates))
    
    task.finished(num_snvs=num_docs, updated_snvs=updates)
```
