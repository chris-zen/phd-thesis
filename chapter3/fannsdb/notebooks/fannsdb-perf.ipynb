```python
%%px --local
def fetch(in_path, out_path, predictors, db_uri, tasks_db_uri=None):
    from fannsdb.db.mongodb import FannsMongoDb
    
    name = os.path.basename(in_path)
    
    if os.path.isdir(out_path):
        in_dir = os.path.dirname(in_path)
        if os.path.realpath(in_dir) == os.path.realpath(out_path):
            name += ".scores"
        out_path = os.path.join(out_path, name)

    if os.path.exists(out_path):
        return
    
    coord_cols = ["protein", "aa_pos", "aa_ref", "aa_alt"]
    
    with Task(tasks_db_uri or db_uri, "fanns-perf-fetch", name) as task:
    
        log = task.logger
    
        log.info("Fetching ...")

        fdb = FannsMongoDb(uri=db_uri)

        with open(in_path) as f, open(out_path, "w") as wf:
            wf.write("\t".join(["ID"] + [c.upper() for c in coord_cols + predictors]) + "\n")

            for num_snvs, line in enumerate(f, start=1):
                fields = line.rstrip("\n").split("\t")
                protein, aa_pos, aa_ref, aa_alt = fields[0:4]
                src_id = protein
                scores = {}
                for d in fdb.query_scores(predictors=predictors, protein=protein,
                                          aa_pos=int(aa_pos), aa_ref=aa_ref, aa_alt=aa_alt):
                    protein = d["protein"]
                    s = d["scores"]
                    for pred, score in s.items():
                        if score is not None and pred not in scores:
                            scores[pred] = score

                # [str(coord[c]) for c in coord_cols]
                wf.write("\t".join([src_id, protein, aa_pos, aa_ref, aa_alt] + [str(scores.get(p, "")) for p in predictors]) + "\n")

                if num_snvs % 1000 == 0:
                    task.progress(num_snvs)

        log.info("[{}] Fetched {} SNVs".format(name, num_snvs))
        
        task.metadata(num_snvs=num_snvs)
        
    return name
```

```python
tset_size = {}
for pos_name, neg_name in tsets:
    name = "{}__{}".format(pos_name, neg_name)
    tset_name = os.path.join(TSETS_PATH, name)
    pos_count = neg_count = 0
    with open(tset_name, "w") as wf:
        with open(os.path.join(SCORES_PATH, pos_name)) as f:
            wf.write(f.readline())
            for line in f:
                pos_count += 1
                i = line.find("\t")
                wf.write("POS" + line[i:])
        with open(os.path.join(SCORES_PATH, neg_name)) as f:
            f.readline()
            for line in f:
                neg_count += 1
                i = line.find("\t")
                wf.write("NEG" + line[i:])
    
    tset_size[name] = { "pos" : pos_count, "neg" : neg_count, "total" : pos_count + neg_count }
    
    print("{:36}  POS: [{:7}]  NEG: [{:7}]".format(name, pos_count, neg_count))
```

```python
%%px --local
def calc_metrics(tset_path, metrics_base_path, predictors, db_uri, overwrite=False):
    import os
    import json
    from condel.metrics import perf_metrics
    
    basename = os.path.basename(tset_path)
    
    metrics_path = os.path.join(metrics_base_path, basename)
    if not overwrite and os.path.exists(metrics_path):
        return
    
    with Task(db_uri, "fanns-perf-metrics", basename) as task:
        task.logger.info("Calculating metrics for {} ...".format(basename))

        #if basename == "cw-2__humvar-n":
        #    m = perf_metrics(tset_path, predictors, transforms=TRANSFORMS, title=basename,
        #                     undersampling=False, scores=PERF_SCORES)
        #else:
        try:
            m = perf_metrics(tset_path, predictors, transforms=TRANSFORMS, title=basename,
                         undersampling=True, num_samples=12, scores=PERF_SCORES)
        except Exception as ex:
            task.logger.exception(ex)
            raise ex
            
        task.logger.info("Saving metrics ...")
        with open(metrics_path, "w") as f:
            json.dump(m, f)

        task.logger.info("Done")
```

```python
%%px --local
def plot_metrics(metrics_path, image_base_path, predictors):
    import os
    import json
    from condel.metrics import plot_perf_metrics
    
    basename = os.path.basename(metrics_path)
    image_path = os.path.join(image_base_path, basename + ".png")
    if os.path.exists(image_path):
        return
    
    logger.info("Loading metrics for {} ...".format(basename))
    
    with open(metrics_path) as f:
        m = json.load(f)
    
    logger.info("Plotting ...")
    
    plot_perf_metrics(m, image_path, predictors=predictors, perf_scores=["ACC", "MCC"])
    
    logger.info("Done")
```

### Create the tables containing the metrics

```python
clean_path(METRICS_TABLES_PATH)
```

```python
import numpy as np
import pandas as pd
import json

num_tset = len(tset_names)
scores = PERF_SCORES + ["AUC", "NUM_POS", "NUM_NEG"]
o = len(PERF_SCORES)

for name in tset_names:
    metrics_path = os.path.join(METRICS_PATH, name)
    basename = os.path.basename(metrics_path)
    logger.info("Loading {} ...".format(basename))
    with open(metrics_path) as f:
        m = json.load(f)["metrics"]
        table = np.zeros((len(ALL_PREDICTORS), len(scores)), dtype=np.float64)
        for i, pred in enumerate(ALL_PREDICTORS):
            pm = m[pred]
            if "perf" not in pm:
                table[i, :] = np.nan
                continue
            perf = pd.DataFrame(pm["perf"], index=pm["cutoffs"])
            best_perf_index = pd.Series(pm["best_perf_index"])
            best_mcc_index = best_perf_index["MCC"]
            ncumhist = pm["neg_cumhist"]
            ncummax = float(ncumhist[-1])
            pcumhist = pm["pos_cumhist"]
            pcummax = float(pcumhist[-1])
            table[i, 0:o] = perf.iloc[best_mcc_index, :][PERF_SCORES]
            table[i, o] = pm.get("auc")
            table[i, o+1] = pcummax
            table[i, o+2] = ncummax
        table = pd.DataFrame(table, columns=scores, index=ALL_PREDICTORS)
        table["POS_CVG"] = table["NUM_POS"] * (100.0 / tset_size[name]["pos"])
        table["NEG_CVG"] = table["NUM_NEG"] * (100.0 / tset_size[name]["neg"])
        table["CVG"] = table[["NUM_NEG", "NUM_POS"]].sum(axis=1) * (100.0 / tset_size[name]["total"])
        table.to_csv(os.path.join(METRICS_TABLES_PATH, basename + ".txt"), sep="\t")
```

```python
def plot_mccs(datasets, image_name, num_cols = 3):
    
    num_ds = len(datasets)
    num_rows = (num_ds + num_cols - 1) / num_cols
    
    #plt.ioff()

    figsize = (6 * num_cols, 5 * num_rows)
    fig, axes = plt.subplots(nrows=(num_ds + num_cols - 1) / num_cols, ncols=num_cols, figsize=figsize)
    plt.subplots_adjust(hspace=0.35)

    tfic_labels = ["T_" + p for p in PREDICTORS]
    is_tfic = [p in TFIC_PREDICTORS for p in tfic_labels]

    bw = np.array([0.45 if p else 0.85 for p in is_tfic])

    x = np.arange(len(PREDICTORS))
    xt = x + 0.45
    yt = np.arange(0, 110, 10)

    for i, ds in enumerate(datasets):
        if ds is None:
            continue

        d = mcc[ds]
        td = [d[p] if p in TFIC_PREDICTORS and p in d else 0.0 for p in tfic_labels]

        ax = axes[i / num_cols, i % num_cols]
        ax.set_title(ds.replace("__", " / "))
        ax.set_ylim(0.0, 100.0)
        ax.set_yticks(yt)
        ax.set_xticks(xt)
        ax.set_xticklabels(PREDICTORS, rotation=90)
        ax.bar(x, d[PREDICTORS], bw, color=colors, alpha=0.9)
        ax.bar(xt, td, bw, color=colors, alpha=0.6)
        ax.grid(axis="y")

    plt.savefig(os.path.join(IMAGES_PATH, image_name), bbox_inches='tight', pad_inches=0.2)
    #plt.close(fig)
```

```python
import random
from fannsdb.db.mongodb import FannsMongoDb
from transfic.blt import load_blt_stats
from transfic.calc import tfic_calc

logger.info("Loading BLT stats ...")

blt_stats = load_blt_stats(BLT_PATH)
feat_stats = blt_stats["ENSP00000269305"]
tfic_src_predictors = [p for p in PREDICTORS if p[2:] in TFIC_PREDICTORS]
upd_pred = {p : tp for p, tp in zip(PREDICTORS_WITH_TFIC, TFIC_PREDICTORS)}

logger.info("Fetching scores and calculating TransFIC ...")

fdb = FannsMongoDb(uri=FANNSDB_URI)

scores = pd.DataFrame(np.empty((len(tp53), len(ALL_PREDICTORS))), index=tp53.index, columns=ALL_PREDICTORS)
scores.ix[:,:] = np.nan
for change in scores.index:
    l = list(fdb.query_scores(predictors=PREDICTORS, protein="ENSP00000269305",
                              aa_pos=int(change[1:-1]), aa_ref=change[0], aa_alt=change[-1]))
    if len(l) == 0:
        continue
    
    d = random.sample(l, 1)[0]
    ds = d["scores"]
    
    tfic = tfic_calc(feat_stats, ds, PREDICTORS_WITH_TFIC, upd_pred, TFIC_TRANSFORMS)
    
    ds.update(tfic)
    
    scores.ix[change] = pd.Series(ds)

logger.info("Done")

display(scores.head())
```

Calculate the mean of activity and the scores for each bin:

```python
def load_scores_and_calc_tfic(muts):
    import random
    from fannsdb.db.mongodb import FannsMongoDb
    from transfic.blt import load_blt_stats
    from transfic.calc import tfic_calc

    logger.info("Loading BLT stats ...")

    blt_stats = load_blt_stats(BLT_PATH)
    
    tfic_src_predictors = [p for p in PREDICTORS if p[2:] in TFIC_PREDICTORS]
    upd_pred = {p : tp for p, tp in zip(PREDICTORS_WITH_TFIC, TFIC_PREDICTORS)}

    logger.info("Fetching scores and calculating TransFIC ...")

    fdb = FannsMongoDb(uri=FANNSDB_URI)

    scores = pd.DataFrame(np.empty((len(muts), len(ALL_PREDICTORS))), index=muts.index, columns=ALL_PREDICTORS)
    scores.ix[:,:] = np.nan
    for index in muts.index:
        mut = muts.ix[index]
        
        feat_stats = blt_stats.get(mut["PROTEIN"])
        if feat_stats is None:
            continue
        
        l = list(fdb.query_scores(predictors=PREDICTORS, protein=mut["PROTEIN"],
                                  aa_pos=mut["POS"], aa_ref=mut["REF"], aa_alt=mut["ALT"]))
        if len(l) == 0:
            continue

        d = random.sample(l, 1)[0]
        ds = d["scores"]

        tfic = tfic_calc(feat_stats, ds, PREDICTORS_WITH_TFIC, upd_pred, TFIC_TRANSFORMS)

        ds.update(tfic)

        scores.ix[index] = pd.Series(ds)

    logger.info("Done")
    
    return scores
```
