## Fetching of scores and training datasets

```python
def load_data(path, *arg, **kwarg):
    data = pd.read_csv(path, *arg, **kwarg)
    display(HTML('<a href="{0}">{0}</a> with <strong>{1}</strong> rows'.format(path, len(data))))
    display(data.head())
    return data
```

Now I will query the **MA web page** for the following mutations and compare the results with the ones obtained from **fannsdb**

```python
d = load_data("tsets/wg-CGC__wg-noCGC", sep="\t", usecols=COLUMNS)
d["MA_WEB"] = np.nan
```

```python
import sys
from time import sleep
import requests as rq

print "Querying for {} mutations ...".format(len(d.index))

with open("valid/ma-web.txt", "w") as f, open("valid/ma-web-err.txt", "w") as ef:
    for index in d.index:
        p = str(d["SWISSPROT_ID"][index])
        var = "{}{}{}".format(d["AA_REF"][index], d["AA_POS"][index], d["AA_ALT"][index])
        q = "http://mutationassessor.org/?cm=var&p={}&var={}&frm=txt".format(p, var)
        try:
            r = rq.get(q)
            text = r.text.replace("\\t", "\t")
            rr = [x.split("\t") for x in text.split("\n")]
            d["MA"][index] = float(d["MA"][index])
            d["MA_WEB"][index] = float(rr[1][7])
            f.write("\t".join(rr[1]))
            sys.stdout.write("\r{} {} {} {} {}       ".format(index, p, var, d["MA"][index], d["MA_WEB"][index]))
            sleep(0.15)
        except Exception as e:
            import traceback as tb
            ef.write("\t".join(rr[1]))
            sys.stdout.write("\r{} Error querying for {} {} --> '{}'".format(index, p, var, rr[1][7]))
            #sys.stdout.write(tb.format_exc() + "\n")
            #sys.stdout.write(text + "\n")
    
    sys.stdout.write(" " * 40 + "\r")

d.to_csv("valid/wg-CGC__wg-noCGC-web", sep="\t", index=False)
```
