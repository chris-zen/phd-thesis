## Retrieve scores for the datasets

```python
!rm scores/*
results = []
for in_path in sorted(glob("snvs/*")):
    out_path = os.path.join("scores", os.path.basename(in_path))
    cmd = "sort -u {} | fannsdb-fetch {} - {} -p {} -a prot_gene,swissprot_id,swissprot_acc".format(in_path, DB_PATH, out_path, PREDICTORS)
    results += [(cmd, out_path, lb.apply_async(check_output, cmd, stderr=STDOUT, shell=True))]

display_results_with_output(results)
```

Filter scores for the longest transcript

```python
!rm scores-filt/*
results = []
for in_path in sorted(glob("scores/*")):
    out_path = os.path.join("scores-filt", os.path.basename(in_path))
    cmd = "transfic-filter-transcripts ../trs_len.tsv {} {} --key ID".format(in_path, out_path)
    results += [(cmd, out_path, lb.apply_async(check_output, cmd, stderr=STDOUT, shell=True))]

display_results_with_output(results)
```

## Prepare training sets

```python
tsets = [
    ("wg-CGC", "wg-noCGC"),
    ("wg-2", "wg-1"),
    ("cosmic-D", "poly-n"),
    ("cosmic-D", "cosmic-O"),
    ("cosmic-CGC", "cosmic-noCGC"),
    ("cosmic-5", "poly-n"),
    ("cosmic-2", "poly-n"),
    ("cosmic-5", "cosmic-1"),
    ("cosmic-2", "cosmic-1")
]

tset_names = ["{}__{}".format(p, n) for p, n in tsets]

!rm tsets/*

for pos_name, neg_name in tsets:
    tset_name = os.path.join("tsets", "{}__{}".format(pos_name, neg_name))
    pos_count = neg_count = 0
    with open(tset_name, "w") as wf:
        with open(os.path.join("scores-filt", pos_name)) as f:
            wf.write(f.readline())
            for line in f:
                pos_count += 1
                wf.write("POS" + line[line.index("\t"):])
        with open(os.path.join("scores-filt", neg_name)) as f:
            f.readline()
            for line in f:
                neg_count += 1
                wf.write("NEG" + line[line.index("\t"):])
                
    display(FileLink(tset_name))
    print "  POS: [{:7}]  NEG: [{:7}]".format(pos_count, neg_count)
```

## Calculate statistics for the training sets

For each training set calculate:

- scores distribution
- cummulative distribution
- MCC and Accuracy
- ROC curve and AUC

The positive and negative data will be balanced using randomizations.

```python
!rm metrics/*

def calc_metrics(predictors, in_path):
    import os
    import json
    from condel.metrics import perf_metrics
    
    prefix = os.path.basename(in_path)
    metrics_path = os.path.join("metrics", "{}.json".format(prefix))

    transforms = dict(
        SIFT=lambda v: 1.0 - v,
        FATHMMC=lambda x: -x)
    
    m = perf_metrics(predictors, in_path, transforms=transforms, title=prefix)
    with open(metrics_path, "w") as f:
        json.dump(m, f)
    
    return metrics_path

results = []
for in_path in [os.path.join("tsets", n) for n in tset_names]:
    results += [lb.apply_async(calc_metrics, PREDICTORS.split(","), in_path)]

for ar in results:
    metrics_path = ar.get()
    display(FileLink(metrics_path))
```

```python
!rm images/*

def plot_metrics(in_path):
    import os
    import json
    from condel.metrics import plot_perf_metrics
    
    prefix = os.path.splitext(os.path.basename(in_path))[0]
    image_path = os.path.join("images", "{}.png".format(prefix))
    
    with open(in_path) as f:
        m = json.load(f)
    
    plot_perf_metrics(m, image_path)
    
    return image_path

results = []
for in_path in [os.path.join("metrics", "{}.json".format(n)) for n in tset_names]:
    results += [lb.apply_async(plot_metrics, in_path)]

for ar in results:
    image_path = ar.get()
    display(FileLink(image_path))
    #display(Image(image_path, retina=True))
```
